```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define datapaths
module_path = os.path.abspath('')
data_path = os.path.join(module_path, 'data')
file_path = os.path.join(data_path, 'products.gz')
```

```python
dataset = pd.read_parquet(file_path, engine='pyarrow')
```

```python
dataset.describe()
```

```python
# unique products
print("Number of Vendors: {}".format(dataset.vendor_id.nunique()))
print("Number of Products: {}".format(dataset.product_id.nunique()))
```
